```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#1. Importing the Data

```python
price = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DA/NYC Airbnb Market/airbnb_price.csv')
review = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DA/NYC Airbnb Market/airbnb_last_review.tsv', sep = '\t')
room_type = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/DA/NYC Airbnb Market/airbnb_room_type.xlsx')
```

```python
display(price.head(), "\n", review.head(), "\n", room_type.head())
```

#2. Cleaning the price column

```python
price['price'] = price['price'].str.strip(" dollars")

price['price'] = price['price'].astype(int)

print(price['price'].describe())
```

```python
free_listing = price[price['price'] == 0]
price = price.drop(free_listing.index)
avg_price = price['price'].mean()
print("The average price per night for an Airbnb listing in NYC is ${:.2f}.".format(avg_price))
```

```python
count_price = sns.displot(x = 'price', data = price)
plt.xscale("log")
plt.title("Log scale histogram of the prices")
plt.show()
```

#4. Comparing costs to the private rental market

```python
price["price_per_month"] = price["price"] * 365 / 12
avg_price_per_month = round(price["price_per_month"].mean(), 2)
print("Airbnb monthly costs are ${}, while in the private market you would pay {}.".format(avg_price_per_month, "$3,100.00"))
```

#5. Cleaning the room type column

```python
room_type["room_type"] = room_type["room_type"].str.lower()
room_type["room_type"] = room_type["room_type"].astype("category")
room_type_count = room_type["room_type"].value_counts()
print(room_type_count)
```

```python
count_room_type = sns.countplot(x = 'room_type', data = room_type, order = room_type_count.index, palette="Pastel1")
count_room_type.bar_label(count_room_type.containers[0])
plt.title("The Counts of different room types")
plt.show()
```

#6. What timeframe are we working with?

```python
review["last_review"] = pd.to_datetime(review["last_review"])
first_review = review["last_review"].dt.date.min()
last_review = review["last_review"].dt.date.max()
print("The latest Airbnb review is {}, the earliest review is {}".format(last_review, first_review))
```

#7. Joining the DataFrames

```python
room_and_price = price.merge(room_type, how = "outer", on = "listing_id")
airbnb_merge = room_and_price.merge(review, how = "outer", on = "listing_id")
airbnb_merge.dropna(inplace = True)
print("There are {} duplicates in the DataFrame.".format(airbnb_merge.duplicated().sum()))
```

#8. Analyzing listing prices by NYC borough

```python
airbnb_merge
```

```python
airbnb_merge["borough"] = airbnb_merge["nbhood_full"].str.partition()[0].str.strip(",")
borough_group = airbnb_merge.groupby("borough")["price"].agg(["sum", "mean", "median", "count"])
borough = borough_group.round(2).sort_values("mean", ascending = False)
print(borough)
```

```python
label_price = ["Budget", "Average", "Expensive", "Extravagant"]
range = [0, 69, 175, 350, np.inf]

airbnb_merge["price_range"] = pd.cut(airbnb_merge["price"], bins = range, labels = label_price)
price_by_borough = airbnb_merge.groupby(["price_range", "borough"])["price_range"].count()
print(price_by_borough)
```

```python
plt.figure(figsize=[10, 10])
price_level = sns.countplot(data = airbnb_merge, x = "borough", hue = "price_range", palette="Pastel1")
for container in price_level.containers:
  price_level.bar_label(container, )
plt.title("Counts of Price levels through the boroughs")
```
